```python
import matplotlib.pyplot as plt
import csv
from scipy import stats
import numpy as np
import sklearn as sl
from sklearn import linear_model
```

```python
# Load the provided data file with the used car data (you can also have a look at it with any text editor)

filename = "data/km_year_power_price.csv"

# place your loading code here

with open(filename, newline='') as f:
    reader = csv.reader(f, delimiter=',')
    """
    the fnction csv.reader() returns a reader object which will iterate over lines in the given csvfile
    
    """
    # get header from first row
    headers = next(reader)
    """
    the function next returns the next row of the readerâ€™s iterable object as a list
    (if the object was returned from reader()) 
    """
    # get all the rows as a list
    data = list(reader)
    """
    the previous line creates a list of lists
    namely, each element of the list is itself a list containing the parameters of each sample (as strings)
    
    """
    # transform data into numpy array
    data = np.array(data).astype(float)
    
# some tests
print(headers)

print("\n")

print(type(data))
print(data[:3].shape)
print(data[:3])

print("\n")

print(type(data[:3,1]))
print(data[:3,1].shape)
print(data[:3,1])

print("\n")

print(type(data[:3,1].reshape(-1,1)))
print(data[:3,1].reshape(-1,1).shape)
print(data[:3,1].reshape(-1,1))
```

Use linear regression to estimate the car prices from the year, kilometers or engine power. 
You can make a simple 1D regression from each one of the parameters independently 




```python
# linear regression with linregress (estimate price from year)

# your code.....

reg = stats.linregress(data[:,1], data[:,3])
"""
the indexing data[:,n] returns the n+1 column of the 2D array
namely, a 1D array containing the (n+1)-th parameter of all the samples

"""
print("slope (linregress):", reg.slope)
print("intercept (linregress):", reg.intercept)
print("correlation coefficient (linregress):", reg.rvalue)
```

```python
# (Optional) linear regression with linear_model.LinearRegression() (estimate price from year)
# Recall that in Python a mx1 matrix is different from a 1D array -> need to reshape

# your code.....

reg = sl.linear_model.LinearRegression().fit(data[:,1].reshape(-1,1), data[:,3].reshape(-1,1))
"""
passing 1D arrays to fit(), i.e. data[:,1], is deprecated
the function fit() requires a mxn matrix, i.e data[:,1].reshape(-1, n), if your data has m samples and n features

"""

print("slope (sl.linear_model):", reg.coef_)
print("intercept (sl.linear_model):", reg.intercept_)

# The coefficient of determination R^2 is equal to the correlation coefficient squared r^2
# only under some very specific conditions
print("correlation coefficient (sl.linear_model):", np.sqrt(reg.score(data[:,1].reshape(-1,1), data[:,3].reshape(-1,1))))
```

```python
# Plot the data and the lines representing the output of the linregress and least square algorithms

# your code....

reg = stats.linregress(data[:,1], data[:,3])

plt.plot(data[:,1], data[:,3], 'o', label='original data')
plt.plot(data[:,1], reg.intercept + reg.slope*data[:,1], 'r', label='fitted line linregress')
plt.legend()
plt.xlabel("year")
plt.ylabel("price [$]")
plt.show()

print("correlation coefficient wrt the year:", reg.rvalue)
```

```python
# linear regression with linregress (estimate price from power)

# your code.....

reg = stats.linregress(data[:,2], data[:,3])

plt.plot(data[:,2], data[:,3], 'o', label='original data')
plt.plot(data[:,2], reg.intercept + reg.slope*data[:,2], 'r', label='fitted line linregress')
plt.legend()
plt.xlabel("power")
plt.ylabel("price [$]")
plt.show()

print("correlation coefficient wrt the power:", reg.rvalue)
```

```python
# linear regression with linregress (estimate price from km)

# your code...

reg = stats.linregress(data[:,0], data[:,3])

plt.plot(data[:,0], data[:,3], 'o', label='original data')
plt.plot(data[:,0], reg.intercept + reg.slope*data[:,0], 'r', label='fitted line linregress')
plt.legend()
plt.xlabel("km")
plt.ylabel("price [$]")
plt.show()

print("correlation coefficient wrt the km:", reg.rvalue)
```

```python
# (Optional) 2D linear regression with linear_model.LinearRegression() (estimate price from year and power)

# your code......

reg = sl.linear_model.LinearRegression().fit(data[:,1:3].reshape(-1,2), data[:,3].reshape(-1,1))
print("slope (sl.linear_model):", reg.coef_)
print("intercept (sl.linear_model):", reg.intercept_)
print("correlation coefficient (sl.linear_model):", np.sqrt(reg.score(data[:,1:3].reshape(-1,2), data[:,3].reshape(-1,1))))

# 3D plot (blue real data, red estimated ones)

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data[:,1], data[:,2], data[:,3], zdir='z', s=5, c='b', depthshade=True)
ax.scatter(data[:,1], data[:,2], reg.intercept_+reg.coef_[:,0]*data[:,1]+reg.coef_[:,1]*data[:,2], zdir='z', s=5, c='r', depthshade=True)
plt.show()
```
